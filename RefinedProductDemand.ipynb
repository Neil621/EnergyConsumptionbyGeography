```python


#read in csv file
reading_in = pd.read_csv('/Users/neilwatt/Documents/BIs/PrWeb/2018Posts/August/Energy/New/ChinaIndia.csv', encoding = "ISO-8859-1")


dfChinaIndia = pd.DataFrame(reading_in, columns=["Year","China Gasoline Demand","China Population","China Gasoline Demand Per Capita","India Gasoline Demand","India Population","India Gasoline Demand Per Capita","U.S Gasoline Demand","U.S Population","U.S Gasoline Demand Per Capita"])
dfChinaIndia
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

title = 'Gasoline Demand Per Capita 2005 to 2017'

labels = ['China', 'India']

colors = ['orange', 'lightblue', 'rgb(49,130,189)', 'rgb(189,189,189)']

mode_size = [8, 8]

line_size = [2, 2]

dates=dfChinaIndia['Year']

#x_data = [dates,dates]

x_data=np.vstack([dates,dates])



demand=dfChinaIndia['China Gasoline Demand Per Capita']
Indiademand=dfChinaIndia['India Gasoline Demand Per Capita']
#USdemand=dfChinaIndia['U.S Gasoline Demand Per Capita']


y_data=np.vstack([demand,Indiademand])
#y_data = [
 #   demand,
  #  Indiademand
#]

traces = []

for i in range(0, 2):
    traces.append(go.Scatter(
        x=x_data[i],
        y=y_data[i],
        mode='lines',
        line=dict(color=colors[i], width=line_size[i]),
        connectgaps=True,
    ))

    traces.append(go.Scatter(
        x=[x_data[i][0], x_data[i][12]],
        y=[y_data[i][0], y_data[i][12]],
        mode='markers',
        marker=dict(color=colors[i], size=mode_size[i])
    ))

layout = go.Layout(
    xaxis=dict(
        showline=True,
        showgrid=False,
        showticklabels=True,
        linecolor='rgb(204, 204, 204)',
        linewidth=2,
        ticks='outside',
        tickcolor='rgb(204, 204, 204)',
        tickwidth=2,
        ticklen=5,
        tickfont=dict(
            family='Arial',
            size=12,
            color='rgb(82, 82, 82)',
        ),
    ),
    yaxis=dict(
        showgrid=False,
        zeroline=False,
        showline=False,
        showticklabels=False,
    ),
    autosize=False,
    margin=dict(
        autoexpand=False,
        l=100,
        r=20,
        t=110,
    ),
    showlegend=False
)

annotations = []

# Adding labels
for y_trace, label, color in zip(y_data, labels, colors):
    # labeling the left_side of the plot
    annotations.append(dict(xref='paper', x=0.05, y=y_trace[0],
                                  xanchor='right', yanchor='middle',
                                  text=label ,
                            
                                  font=dict(family='Arial',
                                            size=16),
                                  showarrow=False))
    # labeling the right_side of the plot
    annotations.append(dict(xref='paper', x=0.95, y=y_trace[12],
                                  xanchor='left', yanchor='middle',
                                  text='{}'.format(y_trace[12]),
                                  font=dict(family='Arial',
                                            size=16),
                                  showarrow=False))
# Title
annotations.append(dict(xref='paper', yref='paper', x=0.0, y=1.05,
                              xanchor='left', yanchor='bottom',
                              text='Gasoline Demand Per Capita (gal per day per capita)',
                              font=dict(family='Arial',
                                        size=20,
                                        color='rgb(37,37,37)'),
                              showarrow=False))
# Source
annotations.append(dict(xref='paper', yref='paper', x=0.5, y=-0.1,
                              xanchor='center', yanchor='top',
                              text='',
                              font=dict(family='Arial',
                                        size=12,
                                        color='rgb(150,150,150)'),
                              showarrow=False))

layout['annotations'] = annotations

fig = go.Figure(data=traces, layout=layout)
#py.iplot(fig, filename='news-source')
plotly.offline.iplot( fig, validate=False )
```
